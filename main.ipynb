```python
import pandas as pd  # pip install pandas openpyxl
import matplotlib.pyplot as plt 
from openpyxl import load_workbook
```

```python
# ---- READ EXCEL ----
def get_data_from_excel():
    df = pd.read_excel(
        io="2024AUG.xlsx",
        engine="openpyxl",
        sheet_name="Sheet1",
        skiprows=0,
        usecols="C:AS",
        nrows=10000,
    )

    return df
```

```python
df = get_data_from_excel()
```

```python
df.info()
```

```python
# Mengganti nama kolom 'DocDate' menjadi 'Date'
df.rename(columns={'DocDate': 'Date'}, inplace=True)
```

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
    # Add 'Month' column to dataframe
df["Month"] = pd.to_datetime(df["Date"]).dt.month
```

```python
df['Month']=df['Date'].dt.strftime('%b') # Short month name, like 'Jan', 'Feb'
```

```python
# Mengganti nama kolom 'Keterangan' menjadi 'Kategori'
df.rename(columns={'Keterangan': 'Kategori'}, inplace=True)
```

```python
# Membuat kolom baru 'Cust.ID' dengan mengambil karakter sebelum spasi pertama
df['Cust.ID'] = df['ItemCode'].str.split(' ').str[0]
```

```python
# Mengganti nama kolom 'Keterangan' menjadi 'Kategori'
df.rename(columns={'Cust.ID': 'Cust'}, inplace=True)
```

```python
# menghapus kolom yg tidak akan digunakan'
df.drop(columns=['NoCard'], inplace=True)
df.drop(columns=['NoBarrelHanger'],inplace=True)
df.drop(columns=['NoBak'],inplace=True)
df.drop(columns=['OK(pcs)'],inplace=True)
df.drop(columns=['Qty(NG)'],inplace=True)
df.drop(columns=['QInspec'],inplace=True)
df.drop(columns=['% NG'],inplace=True)
```

```python
# Mengatur pandas untuk menampilkan semua kolom
pd.set_option('display.max_columns', None)
df
```

```python
# memindahkan posisi kolom Y ke setelah kolom X
# 1. Daftar kolom saat ini
columns = df.columns.tolist()
# 2. Pindahkan kolom 'Y' setelah kolom 'X'
columns.insert(columns.index('X'), columns.pop(columns.index('Y')))
# 3. Atur ulang dataframe dengan urutan kolom baru
df = df.reindex(columns=columns)
```
